```python
for item in soup.select('.sr_property_block'): 
        Hotel_name=  (item.select('.sr-hotel__name')[0].get_text().strip())
        Hotel_link=  (item.select('.hotel_name_link')[0]['href'])
        Hotel_score= (item.select('.bui-review-score__badge')[0].get_text().strip())
        Hotel_reviews=  (item.select('.bui-review-score__text')[0].get_text().strip())
        Hotel_rank= (item.select('.bui-review-score__title')[0].get_text().strip())
        Hotel_price=(item.select('.bui-price-display__value')[0].get_text().strip())
```
